```python
import numpy as np
import matplotlib.pyplot as plt
```

# Creating Constants

```python
G = 6.67430e-11  # Gravitational constant
M = 5.972e24     # Mass of the black hole (in kilograms)
c = 3.0e8        # Speed of light (in m/s)
Rs = 2 * G * M / c**2  # Schwarzschild radius (event horizon)
```

# Drawing Grid for simulation

```python
x = np.linspace(-10, 10, 500)
y = np.linspace(-10, 10, 500)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
```

# Calculating gravitational lensing effect

```python
def lensing_effect(R, Rs):
    return np.where(R < Rs, 0, 1 / np.sqrt(1 - Rs/R))

Z = lensing_effect(R, Rs)
```

# Plotting the gravitational lensing

```python
plt.imshow(Z, extent=(-10, 10, -10, 10), cmap='inferno')
plt.colorbar(label='Lensing Effect (Intensity)')
plt.title("2D Gravitational Lensing around a Black Hole")
plt.xlabel("X Distance (AU)")
plt.ylabel("Y Distance (AU)")
plt.show()
```
